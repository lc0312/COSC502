```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Question 1

# a) Single Variable Numerical

quiz_11 = [28,30,10,30,10,18,30,12,30,8,26,12,30,28,28,10,28,28,28,28,30,30,30]
quiz_22 = [27,27,24,27,30,24,27,21,27,27,30,27,27,30,30,24,21,21,27,24,30,30,24]

quiz_1 = pd.DataFrame (quiz_11)
quiz_2 = pd.DataFrame (quiz_22)

print('Describe for quiz 1', quiz_1.describe())
xran = np.linspace(0,30,30)

print ('Mode for quiz 1', quiz_1.mode().values[0]) 

print ('Histogram for quiz 1')

plt.hist(quiz_1,xran)
plt.show()

print('Describe for quiz 2', quiz_2.describe())

print ('Mode for quiz 2', quiz_2.mode().values[0]) 

print ('Histogram for quiz 2')

plt.hist(quiz_2, xran)
plt.show()


print ('quzi 2 is more consistent becasue it has smaller std')
```

```python
# a) Graphical analysis 

plt.hist ([quiz_11,quiz_22], xran, label=['Quiz 1', 'Quzi 2'])
plt.legend(loc=0) 
plt.show ()

print ('Seems like quiz 2 is more consistent, because the data stay closer with each other')
```

```python
# b) find the students who did really bad in Quiz 1
far_away = quiz_1.describe().values[1]-3*quiz_1.describe().values[2]
print ('3 standard deviation from mean is', far_away)
print ('obviously, this is not real in quzi 1, based on 3 standard rule, no one is really bad, \n because the minimal score is 8, larger than -1.75')
```
