```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Question 1

# a) Single Variable Numerical

quiz_11 = [28,30,10,30,10,18,30,12,30,8,26,12,30,28,28,10,28,28,28,28,30,30,30]
quiz_22 = [27,27,24,27,30,24,27,21,27,27,30,27,27,30,30,24,21,21,27,24,30,30,24]

quiz_1 = pd.DataFrame (quiz_11)
quiz_2 = pd.DataFrame (quiz_22)

print('Describe for quiz 1', quiz_1.describe())
xran = np.linspace(0,30,30)

print ('Mode for quiz 1', quiz_1.mode().values[0]) 

print ('Histogram for quiz 1')

plt.hist(quiz_1,xran)
plt.show()

print('Describe for quiz 2', quiz_2.describe())

print ('Mode for quiz 2', quiz_2.mode().values[0]) 

print ('Histogram for quiz 2')

plt.hist(quiz_2, xran)
plt.show()


print ('quzi 2 is more consistent becasue it has smaller std')
```

```python
# Qeestion 4
# 1) 
import pandas as pd

url = 'https://github.com/lc0312/COSC502/blob/main/iris.csv?raw=true'

df = pd.read_csv(url)
print ('data points have shape of', df.shape)
print ('In total is 600')
```

```python
# 2)
print ('name of each col and dtypes are:')
print (df.dtypes)
```

```python
# 3)
s_3 = df['species'].nunique()

print ('the number of species is:', s_3)
```

```python
# 4)
df_species = df['species'].unique()

print ('species are:', df_species)

df_s0 = df[df['species'] == 'setosa']

df_s0_sr = df_s0['sepal_length'] / df_s0['sepal_width'] # sepal ration for setosa
df_s0_pr = df_s0['petal_length'] / df_s0['petal_width'] # petal ration for setosa

print ('for setosa sepal')
print (df_s0_sr.agg(['mean','median', 'min', 'max', 'std']))
print ('for setosa petal')
print (df_s0_pr.agg(['mean','median', 'min', 'max', 'std']))

df_s1 = df[df['species'] == 'versicolor']

df_s1_sr = df_s1['sepal_length'] / df_s1['sepal_width'] # sepal ration for setosa
df_s1_pr = df_s1['petal_length'] / df_s1['petal_width'] # petal ration for setosa

print ('for versicolor sepal')
print (df_s1_sr.agg(['mean','median', 'min', 'max', 'std']))
print ('for versicolor petal')
print (df_s1_pr.agg(['mean','median', 'min', 'max', 'std']))

df_s2 = df[df['species'] == 'virginica']

df_s2_sr = df_s2['sepal_length'] / df_s2['sepal_width'] # sepal ration for setosa
df_s2_pr = df_s2['petal_length'] / df_s2['petal_width'] # petal ration for setosa

print ('for virginica sepal')
print (df_s2_sr.agg(['mean','median', 'min', 'max', 'std']))
print ('for virginica petal')
print (df_s2_pr.agg(['mean','median', 'min', 'max', 'std']))
```

```python
# 5）
import matplotlib.pyplot as plt

df.loc[df.species == 'setosa', 's_ratio'] = df_s0_sr
df.loc[df.species == 'setosa', 'p_ratio'] = df_s0_pr

df.loc[df.species == 'versicolor', 's_ratio'] = df_s1_sr
df.loc[df.species == 'versicolor', 'p_ratio'] = df_s1_pr

df.loc[df.species == 'virginica', 's_ratio'] = df_s2_sr
df.loc[df.species == 'virginica', 'p_ratio'] = df_s2_pr

boxplot_0 = df.boxplot (column=['s_ratio'], by='species')
boxplot_0.set_ylabel ('s_ratio')
boxplot_0.set_title ('sepal ratio')
boxplot_1 = df.boxplot (column=['p_ratio'], by='species')
boxplot_1.set_ylabel ('p_ratio')
boxplot_0.set_title ('petal ratio')

plt.show()


```

```python
# 6)
df.loc[df.species == 'setosa', 'color'] = 'red'
df.loc[df.species == 'versicolor', 'color'] = 'green'
df.loc[df.species == 'virginica', 'color'] = 'blue'

a= df.loc[df.species == 'setosa'].plot.scatter(x="sepal_width", y="sepal_length", c='color', label='setosa')
b= df.loc[df.species == 'versicolor'].plot.scatter(x="sepal_width", y="sepal_length", c='color', label='versicolor', ax=a)
c= df.loc[df.species == 'virginica'].plot.scatter(x="sepal_width", y="sepal_length", c='color', label='virginica', ax=b)

plt.show()

a= df.loc[df.species == 'setosa'].plot.scatter(x="petal_width", y="petal_length", c='color', label='setosa')
b= df.loc[df.species == 'versicolor'].plot.scatter(x="petal_width", y="petal_length", c='color', label='versicolor', ax=a)
c= df.loc[df.species == 'virginica'].plot.scatter(x="petal_width", y="petal_length", c='color', label='virginica', ax=b)

plt.show()
```

```python
# 7)
print ('the ratio of petal (sepal) length to width show clear difference, which can be see in scatter figuer')
print ('the mean value of ratio also tell us we can use this difference to distingush the different plows')
```
